```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist

# The data has already been sorted into training and test sets for us
(train_data, train_labels), (test_data, test_labels) = fashion_mnist.load_data()
```

```python
train_data_norm = train_data / 255.0
test_data_norm = test_data / 255.0
```

```python
model_3 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(28, activation=tf.keras.activations.relu),
    # tf.keras.layers.Dense(40, activation=tf.keras.activations.relu),
    tf.keras.layers.Dense(10, activation = tf.keras.activations.softmax)
])

model_3.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics=['accuracy'])

model_3.fit(train_data_norm, train_labels, epochs=10, validation_data=(test_data_norm, test_labels))
```

```python
# Implement confusion matrix

import itertools
from sklearn.metrics import confusion_matrix
import numpy as np

def make_confusion_matrix (y_true, y_pred, classes=None, figsize=(10,10), text_size=5):
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] # normalize our confusion matrix
  n_classes = cm.shape[0]

  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap = plt.cm.Blues)
  fig.colorbar(cax)

  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  ax.set(title='Confusion Matrix',
         xlabel='Predicted label',
         ylabel='True label',
         xticks=np.arange(n_classes),
         yticks=np.arange(n_classes),
         xticklabels=labels,
         yticklabels=labels)

  # Set x=axis labels to the bottom
  ax.xaxis.set_label_position('bottom')
  ax.xaxis.tick_bottom()

  # Adsjust label size
  ax.yaxis.label.set_size(text_size)
  ax.xaxis.label.set_size(text_size)
  ax.title.set_size(text_size)

  threshold = (cm.max() + cm.min()) / 2

  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j] * 100:.1f}%)",
    horizontalalignment='center',
    color = 'white' if cm[i, j] > threshold else 'black',
    size=text_size)
```

```python
# Make a prettier confusion matrix
make_confusion_matrix(y_true=test_labels,
                      y_pred=y_preds,
                      classes=class_names,
                      figsize=(15, 15),
                      text_size=10
)
```

```python
# Recreate Softmax function
import numpy as np

def softmax(x):
  """
  Compute the softmax of each element along the last axis of x.

  Parameters:
  x (np.ndarray): Input tensor.

  Returns:
  np.ndarray: Tensor with the same shape as x, with the softmax applied to the last axis.
  """
  exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
  return exp_x / np.sum(exp_x, axis = -1, keepdims=True)
```

```python
# Create a function to plot multiple predictions at a time
import random
import matplotlib.pyplot as plt

def plot_random_predictions(model, images, true_labels, classes):
  for j in range(4):
    ax = plt.subplot(2, 2, j + 1)
    i = random.randint(0, len(images))

    # Create predictions
    target_image = images[i]
    pred_probs = model.predict(target_image.reshape(1, 28, 28))
    pred_label = classes[pred_probs.argmax()]
    true_label = classes[true_labels[i]]

    # Plot the image
    plt.imshow(target_image, cmap = plt.cm.binary)

    # Change the color depending if title is right or wrong
    if pred_label == true_label:
      color = 'green'
    else:
      color = 'red'
    plt.title(pred_label, color=color)
    plt.axis(False)
    # plt.imshow(train_data[rand_index], cmap=plt.cm.binary)
    # plt.title(class_names[train_labels[rand_index]])
```

```python
plot_random_predictions(model=model_3, images=test_data_norm, true_labels = test_labels, classes=class_names)
```

```python
# Make a function to show an image of a certain class and make a prediction on it.

def find_label_by_name(names, label):
  for i, title in enumerate(names):
    if title.lower() == label.lower():
      return i
  return None

def show_images_with_predictions(model, images, true_labels, classes, target_class):
  target_label = find_label_by_name(classes, target_class)
  indices = [i for i, label in enumerate(test_labels) if label==target_label]
  for i in range(9):
    ax = plt.subplot(3, 3, i+1)
    index = random.randint(0, len(indices))
    image = images[indices[index]]
    prediction = model.predict(image.reshape(1, 28, 28))
    pred_label = classes[prediction.argmax()]
    true_label = classes[true_labels[indices[index]]]
    plt.imshow(image, cmap=plt.cm.binary)
    if pred_label == true_label:
      color = 'green'
    else:
      color = 'red'
    plt.title(pred_label, color=color)
    plt.axis(False)

show_images_with_predictions(model_3, test_data, test_labels, class_names, 'Bag')
```
